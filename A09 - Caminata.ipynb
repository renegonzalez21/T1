```python
personas = 1000
n = 10
moneda  = ["aguila", "sello"] 
cams = []
for p in range(personas):
    pos = 0
    cam = [0]
    for i in range (n):
        lanzamiento = random.choice (moneda)
        if lanzamiento == "aguila" :
            pos += 1
        else:
            pos -= 1
        cam.append(pos)
    cams.append(cam)
    
```

```python
import matplotlib.pyplot as plt
plt.figure()
for p in range(personas):
    plt.plot(cams[p])
```

```python
import numpy as  np

X = np.array(cams)
(X[:,-1] == 0).mean()

for i in [-10, - 8, -6, -4, 0, 2, 6, 8, 10]:
    print((X[:, -1]==i).mean())
```
