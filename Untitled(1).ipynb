**OBJETIVO**
Maximizar la ganacia total de la carpinteria durante una semana

**Función de costo**
La función a maximizar es la ganacia total obtenida de la venta de las mesas, sillas, libreros y bases de cama . Ganancia total = 9000 * mesas + 3000 * Sillas + 10000 * Libreros + 9000 * Bases de cama

**Variables de Desición**
* $x_1$: Número de mesas producidas
* $x_2$: Numero de sillas producidas
* $x_3$: Numero de libreros producidos
* $x_4$: Númro de  bases de cama
* $y_1$: Número de mesas producidas aprendiz
* $y_2$: Numero de sillas producidas aprendiz
* $y_3$: Numero de libreros producidos aprendiz
* $y_4$: Número de  bases de cama aprendiz
**Restricciones**
1. **Tiempo de Trabajo Empleado:**
* Empleado Experimentado:
    * Tiempo total disponible: 40 horas/semana.
    * Tiempo requerido: 3$x_1$ + $x_2$ + 4$x_3$ + 3$x_4$  $\leq$ 40 horas
* Aprendiz:
    * Tiempo total disponible: 40 horas/semana.
    * Tiempo requerido: 4.5$y_1$ + 1.5$y_2$ + 6$y_3$ + 4.5$y_4$  $\leq$ 40 horas

2. **Materiales disponibles**
   * Unidades de madera: 100 unidades
   * Material requerido: 6$x_1$+2$x_2$+5$x_3$+5$X_4$+6$y_1$+2$y_2$+5$y_3$+5$y_4$ $\leq$ 100 unidades

**Ganancias por producto**
* Mesas: $9000
   
* Sillas: $3000

* Libreros: $10000

* Bases de cama: $9000

**Función Objetivo:** Maximizar la ganancia total: 

Ganancia Total = 9000($x_1 + y_1$) + 3000($x_2 + y_2$) + 10000($x_3 + y_3$) + 9000($x_4 + y_4)

**Información General**
* Tiempo total de trabajo semanal por persona: 40 horas
* Material total disponible: 100 unidades de madera 

```python
import scipy.optimize as opt
import numpy as np
c = np.array([-9000, -3000, -10000, -9000, -9000, -3000, -10000, -9000])

A_ub = np.array([
    [3, 1, 4, 3, 0, 0, 0, 0],        
    [0, 0, 0, 0, 4.5, 1.5, 6, 4.5], 
    [6, 2, 5, 5, 6, 2, 5, 5]       
])

b_ub = np.array([
    40,  
    40,  
    100  
])

sol = opt.linprog(c, A_ub, b_ub,)
sol
```
