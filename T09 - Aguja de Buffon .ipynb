El problema de la aguja de Buffon es un famoso experimento de probabilidad propuesto por el matemático francés Georges-Louis Leclerc, conocido como el conde de Buffon, en el siglo XVIII. Este experimento implica lanzar una aguja de longitud L sobre un plano que tiene líneas paralelas, las cuales están separadas por una distancia **D**. Es importante que **L** sea menor o igual a **D**.

### Explica la solución con cálculo integral usando tus palabras.

Para determinar la probabilidad de que la aguja cruce una línea mediante cálculo integral, consideramos diversos factores:

1. **Ángulo de Caída**
   - La aguja puede caer en un ángulo θ con respecto a las líneas, el cual puede variar entre 0 y π/2.
  
2. **Posición**
   - La distancia desde el centro de la aguja a la línea más cercana se llama x, y puede fluctuar entre 0 y D/2.

La probabilidad de que la aguja cruce una línea se calcula integrando sobre todas las posibles posiciones y ángulos, lo que da lugar a la siguiente fórmula:

$$P = \frac{2L}{\pi D}$$

Esto se deriva al calcular la fracción de la superficie total en el espacio de configuración que resulta en un cruce de línea, considerando las posiciones y los ángulos.

### Solución sin cálculo integral

Sin usar cálculo integral, podemos entender la solución de manera más intuitiva. Imagina que lanzamos la aguja varias veces. 

1. **Posiciones aleatorias**: Al caer la aguja, puede caer en diferentes posiciones respecto a las líneas. Cuanto más cerca esté el centro de la aguja de una línea, mayor será la probabilidad de que la aguja cruce esa línea.
  
2. **Longitud de la aguja y separación de líneas**: Si la longitud de la aguja es similar o menor que la distancia entre las líneas, hay más oportunidades para que la aguja cruce. Si _L_ es muy pequeña comparada con _D_, entonces será menos probable que cruce.

3. **Experimentos repetidos**: Si repetimos el experimento muchas veces y contamos cuántas veces la aguja cruza las líneas, podemos estimar la probabilidad. Esta relación resulta ser proporcional a la longitud de la aguja y la distancia entre líneas, lo que lleva a la fórmula:

Se puede hacer,contando las veces las agujas cruzan las lineas, y posteriormente las distancias entre una y otra, calculandp tambien el angulo en el que cae.


Crea una simulación con numpy donde se arrojen n agujas de longitud l dado que se tiene una distancia d entre cada línea paralela.

    ¿Qué variables aleatorias hay en este caso?
    ¿Qué distribución siguen?
    ¿Cómo evoluciona el resultado contra la cantidad de agujas que se usa?

```python
import numpy as np
import matplotlib.pyplot as plt

def simular_buena_nueva(n, L, D):
    x = np.random.uniform(0, D / 2, n)  
    theta = np.random.uniform(0, np.pi / 2, n)
    
    h = (L / 2) * np.sin(theta)  
    cruces = np.sum(h >= x)  
    return cruces


L = 1.0 
D = 1.0  
n_values = np.arange(10, 10001, 100) 
resultados = []

for n in n_values:
    cruces = simular_buena_nueva(n, L, D)
    resultados.append(cruces / n) 

p_estimado = (2 * L) / (np.pi * D)

plt.figure(figsize=(10, 6))
plt.plot(n_values, resultados, label='Estimación de Probabilidad')
plt.axhline(y=p_estimado, color='r', linestyle='--', label='Probabilidad Teórica')
plt.title('Estimación de la Probabilidad de Cruce de Agujas de Buffon')
plt.xlabel('Número de Agujas')
plt.ylabel('Probabilidad Estimada')
plt.legend()
plt.grid()
plt.show()
```
