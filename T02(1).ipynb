1. Graficar las siguientes funciones f(x). Determina el dominio a partir de la función:

    

1. $f(x) = x^2-4x+2$ 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) 
y = x**2 - 4*x + 2

plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = x^2 - 4x + 2$", color ="b")
plt.title("Gráfica de $f(x) = x^2 - 4x + 2$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend
plt.grid()
plt.show()
```

2. $f(x) = sen(3x)$

```python
x = np.linspace(-1 * np.pi, 1 * np.pi, 100)
y = np.sin(3 * x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = \\sin(3x)$", color="b")
plt.title("Gráfica de $f(x) = \\sin(3x)$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

3. $e^{2x} - 3$

```python
x = np.linspace(0, 2, 100)
y = np.exp(2 * x) - 3
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x) = e^{2x} - 3$", color="b")
plt.title("Gráfica de $f(x) = e^{2x} - 3$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

4. $f(x)= \sqrt{(x^2 - 4)}$

```python
x = np.linspace(0, 10, 100)
y = np.sqrt(x**2 - 4)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="$f(x)=\\sqrt{x^2 - 4}$", color="b")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función $f(x) = \\sqrt{x^2 - 4}$")
plt.grid()
plt.legend()
plt.show()
```

5. $f(x) = cos(x + \frac{\pi}{4})$

```python
x = np.linspace(-1 * np.pi, 1 * np.pi, 400)
y = np.cos(x + np.pi / 4)
plt.figure(figsize=(8, 6))
plt.plot(x, y, color="b", label="$f(x) = \\cos(x + \\frac{\\pi}{4}$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de la función $f(x) = \\cos(x + \\frac{\\pi}{4}$")
plt.grid()
plt.legend()
plt.show()
```

2. Graficar los siguientes grupos de funciones:






1. __Grupo 1__

    $y_1 = 2x + 1$

    $y_2 = -x + 4$

    $y_3 = x^2$

```python
x = np.linspace(-5, 5, 100)
y1 = 2 * x + 1
y2 = -x + 4
y3 = x**2
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label="$y_1 = 2x + 1$", color="b")
plt.plot(x, y2, label="$y_2 = -x + 4$", color="g")
plt.plot(x, y3, label="$y_3 = x^2$", color="r")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de las ecuaciones $y_1 = 2x + 1$, $y_2 = -x + 4$, $y_3 = x^2$")
plt.legend()
plt.grid()
plt.show()
```

2. __Grupo 2__

   $y_1 = \sin(x)$

   $y_2 = \cos(x)$

```python
x = np.linspace(-5, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label="$y_1 = \\sin(x)$", color="b")
plt.plot(x, y2, label="$y_2 = \\cos(x)$", color="g")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de las ecuaciones $y_1 = \\sin(x)$, $y_2 = \\cos(x)$")
plt.legend()
plt.grid()
plt.show()
```

3. Crea una función que cree una gráfica y dibuje 'k' líneas de forma  La función recibe:
    * x1: valor inicial de 'x'
    * x2: valor final de 'x'
    * n: cantidad de puntos del vector 'x'
    * a: valor inicial de image.png
    * b: valor final de image.png
    * k: cantidad de líneas

Usar el la variable opcional de "label" y el comando de legend para saber qué línea pertenece a qué $\alpha$

E.g.resultaría en:



```python
x1 = 0
x2 = 2 * np.pi
n = 1000
a1 = 1
a2 = 3
k = 3
x = np.linspace(x1, x2, n)
plt.figure(figsize=(8, 6))
for i in range(k):
    a = a1 + i * (a2 - a1) / (k - 1)
    y = np.sin(a * x)
    plt.plot(x, y, label= f'a = {a:.1f}')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfica de $f(x) = \\sin(ax)$ con diferentes valores de α")
plt.legend()
plt.grid()
plt.show()
```
